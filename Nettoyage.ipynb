Maintenant pour seconde étape de traitement, on va s'interesser aux ***anomalies*** pour les variables numériques , ***valeurs étrangéres*** pour les variables catégorielles et ***de Dates Anormales*** pour les dates

- Pour la table ***CONTACT***

- Détection d'Outliers avec l'IQR (Intervalle Interquartile)

```python
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[column] < (Q1 - 1.5 * IQR)) | (df[column] > (Q3 + 1.5 * IQR))]
    return outliers

outliers_ID_INDIVIDUS = detect_outliers_iqr(df_contact_cleaned, 'ID_INDIVIDU')
print(outliers_ID_INDIVIDUS)
```
